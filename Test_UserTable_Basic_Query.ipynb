```python
# get wikipedia stuff 
import pymongo
from pymongo import MongoClient
import pandas as pd
import dns
from dns import resolver 

import wikipedia

pd.set_option('display.max_colwidth', None)

client = pymongo.MongoClient("mongodb+srv://AtlasTwitter:1FineTwitterApp!@twittercluster.ycq9k.mongodb.net/")
mongo_db = client["Tweets_DB"]
mongo_collection = mongo_db["Users_Table"]

users_df = pd.DataFrame(list(mongo_collection.find(  )))

users_df
```

```python
# get wikipedia stuff 

def GetWikiSummary(InputUser):
    UserSummary = wikipedia.summary(InputUser, sentences=3)
    return UserSummary
        
# exploded_group_df.reset_index(inplace = True)
users_with_wiki_df = users_df.copy()
users_with_wiki_df['WikiSummary'] = users_df['name'].apply(GetWikiSummary)
users_with_wiki_df
```
