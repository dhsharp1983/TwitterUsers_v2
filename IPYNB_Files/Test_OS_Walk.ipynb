```python
# remove PDF files from CSV folder 
import os
from fnmatch import fnmatch

for dirpath, dirnames, filenames in os.walk("../Original_CSVs"):
    for file in filenames:
        if fnmatch(file, '*.pdf'):
            print("Removing Dir: " + dirpath + "& File: " + file)
            os.remove(os.path.join(dirpath, file))
```
