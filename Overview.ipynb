```python
#Import Dependencies
import requests
import wikipedia
import pandas as pd
from config4 import api_key
```

```python
#Create reference to CSV file
csv_path = "UsersWithWiki.csv"

#Import the CSV into a pandas dataframe
Tweets_df = pd.read_csv(csv_path, low_memory = False)

#Print the dataframe
Tweets_df
```

```python
#Show the 'shape' of the dataframe 
shape = Tweets_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Show the Identity column as a list
list(Tweets_df['Identity'])
```

```python
#Simply change some of the results of the identity to match other components
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['BBC Breaking News'],'BBC')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['CNN Breaking News'],'CNN Breaking')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['Donald J. Trump'],'Donald Trump')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['Drake (Drizzy?)'],'Drake')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['Ellen DeGeneres'],'The Ellen Show')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['Jennifer Lopez'],'J Lo')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['Kim Kardashian West'],'Kim Kardashian')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['LeBron James'],'Le Bron James')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['Lil Wayne WEEZY F'],'Lil Wayne')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['Miley Ray Cyrus'],'Miley Cyrus')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['P!nk'],'Pink')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['Real Madrid C.F.⚽'],'Real Madrid')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['SportsCenter'],'Sports Center')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['The New York Times'],'NY Times')
Tweets_df['Identity'] = Tweets_df['Identity'].replace(['YouTube'],'Youtube')

#Print the dataframe
Tweets_df
```

```python
#Read Excel sheet and convert to dataframe
Sentiment_df = pd.read_excel('Overall Sentiment.xlsx')

#Print the dataframe
Sentiment_df
```

```python
#Rename the column for the purposes of merging the 2 dataframes
Tweets_df = Tweets_df.rename(columns = {"User Name":"Username"})

#Merge (Outer Join) the 2 dataframes
Tweets_df = pd.merge(Tweets_df, Sentiment_df, how='outer', on=['Username', 'Username'])

#Rename the column back
Tweets_df = Tweets_df.rename(columns = {"Username":"User Name"})

#Print the dataframe
Tweets_df
```

```python
#Drop unneccessary columns
Tweets_df = Tweets_df.drop(['Overall sentiment of the user', 'Assigned color for the sentiement (in hexidecimal)'], axis = 1)
```

```python
#Read Excel sheet and convert to dataframe
Residence_df = pd.read_excel('TweetersResidence.xlsx')

#Print the dataframe
Residence_df
```

```python
#Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"

#Set to imperial to gather the data values in farenheit and mph (not relevant to this project)
units = "imperial"

# Build partial query URL
pquery_url = f"{url}appid={api_key}&units={units}&q="
```

```python
#Select the 'Current Residence' column and convert it into a list 
homes = Residence_df['Current Residence'].to_list()

#Confirm the length of the list
len_homes = len(homes)
print(f"The length of the Identities list is: {len_homes} ")
```

```python
#Create empty list
lat = []
lon = []
```

```python
#Loop through the list of Current Residences to populate the latitude lists
for home in homes:
    response = requests.get(pquery_url + home ).json()
    try: 
        #Populate the latitude list with the result from specific point in the URL
        lat.append(response['coord']['lat'])       
        #Print to confirm the latitude value has been updated to the list
        print(f"{home}'s latitude found! Appending stats")     
    except:
        #If there is no data, replace with the value = "None"
        lon.append(None)       
        #Print to inform that there was no data found
        print("Data not found")
```

```python
#Loop through the list of Current Residences to populate the longitude lists
for home in homes:
    response = requests.get(pquery_url + home ).json()
    try: 
        #Populate the longitude list with the result from specific point in the URL
        lon.append(response['coord']['lon'])       
        #Print to confirm the longitude value has been updated to the list
        print(f"{home}'s longitude found! Appending stats")     
    except:
        #If there is no data, replace with the value = "None"
        lon.append(None)       
        #Print to inform that there was no data found
        print("Data not found")
```

```python
#Print the results of the loop
print(f"The latitude information received is: {lat}")
print("===========================================================================================")
print(f"The longitude information received is: {lon}")
print("===========================================================================================")

#Confirm the length of the each list
len_lat = len(lat)
len_lon = len(lon)

#Print the results of the list length
print(f"The length of the Latitude list is: {len_lat} ")
print(f"The length of the Longitude list is: {len_lon} ")
```

```python
#Add column called 'Latitude', and a column titled 'Longitude'
Residence_df['Latitude'] = lat
Residence_df['Longitude'] = lon

#View new data frame
Residence_df
```

```python
#Merge (Outer Join) the 2 dataframes
Overview_df = pd.merge(Tweets_df, Residence_df, how='outer', on=['Identity', 'Identity'])

#Print the dataframe
Overview_df
```

```python
#Select the 'Tweet_Positivity_Rank' column and convert it into a list 
rank = Overview_df['Tweet_Positivity_Rank'].to_list()

#Create a new column titled 'Rank'
Overview_df['Rank'] = rank

#Print the dataframe
Overview_df
```

```python
#Sort the 'Rank' column in decending order
Overview_df.sort_values(by=['Rank'], inplace=True, ascending=False)

#Set the Rank column as the index
Overview_df = Overview_df.set_index('Rank')

#Print the dataframe
Overview_df
```
